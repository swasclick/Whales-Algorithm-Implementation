```python
attack_map = attack_map = {
    # DoS Attacks
    'back': 'DoS',
    'land': 'DoS',
    'neptune': 'DoS',
    'pod': 'DoS',
    'smurf': 'DoS',
    'teardrop': 'DoS',
    'mailbomb': 'DoS',
    'apache2': 'DoS',
    'processtable': 'DoS',
    'udpstorm': 'DoS',

    # R2L Attacks
    'ftp_write': 'R2L',
    'guess_passwd': 'R2L',
    'imap': 'R2L',
    'multihop': 'R2L',
    'phf': 'R2L',
    'spy': 'R2L',
    'warezclient': 'R2L',
    'warezmaster': 'R2L',
    'sendmail': 'R2L',
    'named': 'R2L',
    'snmpgetattack': 'R2L',
    'snmpguess': 'R2L',
    'worm': 'R2L',
    'xlock': 'R2L',
    'xsnoop': 'R2L',

    # U2R Attacks
    'buffer_overflow': 'U2R',
    'loadmodule': 'U2R',
    'perl': 'U2R',
    'rootkit': 'U2R',
    'httptunnel': 'U2R',
    'ps': 'U2R',
    'sqlattack': 'U2R',
    'xterm': 'U2R',

    # Probe Attacks
    'ipsweep': 'Probe',
    'nmap': 'Probe',
    'portsweep': 'Probe',
    'satan': 'Probe',
    'mscan': 'Probe',
    'saint': 'Probe',

    # Normal
    'normal': 'Normal'
}

```

```python

df_train['Attack'] = df_train['Attack'].map(attack_map)
df_test['Attack'] = df_test['Attack'].map(attack_map)
```

```python
df_train.head()
```

```python
columns_to_drop_from_X = ['Attack', 'difficulty_level' , 'num_outbound_cmds']

# Attack is target variable
# 42 is difficulty level , which is metadata of instance itself and not the unique feature
# 20 is almost always has a constant value (usually 0) across all records.

X_train = df_train.drop(columns = columns_to_drop_from_X, axis = 1)
X_test = df_test.drop(columns = columns_to_drop_from_X, axis = 1)

y_train = df_train['Attack']
y_test = df_test['Attack']

```

```python
categorical_features = ['protocol_type', 'service', 'flag']
numerical_features = X_train.select_dtypes(include=['number']).columns.tolist()
numerical_features = [f for f in numerical_features if f not in categorical_features]
```

```python
print(f"\nCategorical features identified: {categorical_features}")
print(f"Numerical features identified: {numerical_features}")
```

```python
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()

print("Numerical features:", numerical_features)
print("Categorical features:", categorical_features)
```

```python
print(f"Shape of X_test after initial drops: {X_test.shape}") # This should be 22544 rows based on your error
print(f"Shape of y_test after initial drops: {y_test.shape}")
```

```python
numerical_transformer = Pipeline(steps = [
    ('scaler', MinMaxScaler())
])
```

```python
categorical_trasformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder = 'passthrough'
)
```

```python
label_encoder= LabelEncoder()
```

```python
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
print(f"Original unique attack categories in y_train: {y_train.unique()}")
print(f"Encoded attack categories mapping: {list(label_encoder.classes_)}")
print(f"First 5 encoded labels in y_train : {y_train_encoded[:5]}")
```

```python

X_train_processed = preprocessor.fit_transform(X_train)
```

```python
print(f"Shape of X_train after preprocessing: {X_train_processed.shape}")
```

```python
print(f"\nClass distribution BEFORE SMOTE on y_train: {Counter(y_train_encoded)}")
```

```python
smote = SMOTE(random_state=42)
```

```python
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_processed, y_train_encoded)
```

```python
print(f"Class distribution AFTER SMOTE on y_train: {Counter(y_train_resampled)}")
```

```python
print(f"Shape of X_train after SMOTE: {X_train_resampled.shape}") # Number of rows will increase here
print(f"Shape of y_train after SMOTE: {y_train_resampled.shape}")
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn_classifier.fit(X_train_resampled, y_train_resampled)
```

```python
X_test_processed = preprocessor.transform(X_test)
```

```python
print(f"Shape of X_test after preprocessing: {X_test_processed.shape}")
```

```python
y_pred_encoded = knn_classifier.predict(X_test_processed)
```

```python
print("--- Model Evaluation ---")
print("Confusion Matrix:")
print(confusion_matrix(y_test_encoded, y_pred_encoded))
print("\nClassification Report:")
print(classification_report(y_test_encoded, y_pred_encoded, target_names=label_encoder.classes_))
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f"\nOverall Accuracy: {accuracy:.4f}")
```
